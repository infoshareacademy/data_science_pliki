```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parametry rozkÅ‚adu normalnego
mu = 125  # Åšrednia (wartoÅ›Ä‡ oczekiwana)
sigma = 15  # Odchylenie standardowe

# Tworzenie instancji rozkÅ‚adu normalnego
distribution = norm(loc=mu, scale=sigma)

# Zakres wartoÅ›ci [Î¼âˆ’4Ïƒ; Î¼+4Ïƒ]
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)

# GÄ™stoÅ›Ä‡ rozkÅ‚adu prawdopodobieÅ„stwa (PDF) i dystrybuanta (CDF)
pdf = distribution.pdf(x)
cdf = distribution.cdf(x)

# Rysowanie wykresÃ³w
plt.figure(figsize=(12, 6))

# GÄ™stoÅ›Ä‡ rozkÅ‚adu prawdopodobieÅ„stwa
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label="GÄ™stoÅ›Ä‡ rozkÅ‚adu (PDF)")
plt.title("GÄ™stoÅ›Ä‡ rozkÅ‚adu prawdopodobieÅ„stwa")
plt.xlabel("WartoÅ›Ä‡")
plt.ylabel("GÄ™stoÅ›Ä‡")
plt.legend()

# Dystrybuanta
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label="Dystrybuanta (CDF)", color="orange")
plt.title("Dystrybuanta rozkÅ‚adu")
plt.xlabel("WartoÅ›Ä‡")
plt.ylabel("PrawdopodobieÅ„stwo")
plt.legend()

plt.tight_layout()
plt.show()

# Obliczanie prawdopodobieÅ„stwa, Å¼e zmienna losowa osiÄ…gnie wartoÅ›Ä‡ z zakresu 130 do 145
probability = distribution.cdf(145) - distribution.cdf(130)

# WartoÅ›Ä‡ oczekiwana
expected_value = distribution.mean()

# Odchylenie standardowe
standard_deviation = distribution.std()

# Kwantyl rzÄ™du 0.25
quantile_25 = distribution.ppf(0.25)

# WyÅ›wietlenie wynikÃ³w
results = {
    "PrawdopodobieÅ„stwo (130 â‰¤ X â‰¤ 145)": probability,
    "WartoÅ›Ä‡ oczekiwana": expected_value,
    "Odchylenie standardowe": standard_deviation,
    "Kwantyl rzÄ™du 0.25": quantile_25,
}

results
```
