```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parametry rozkadu normalnego
mu = 125  # rednia (warto oczekiwana)
sigma = 15  # Odchylenie standardowe

# Tworzenie instancji rozkadu normalnego
distribution = norm(loc=mu, scale=sigma)

# Zakres wartoci [4; +4]
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)

# Gsto rozkadu prawdopodobiestwa (PDF) i dystrybuanta (CDF)
pdf = distribution.pdf(x)
cdf = distribution.cdf(x)

# Rysowanie wykresw
plt.figure(figsize=(12, 6))

# Gsto rozkadu prawdopodobiestwa
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label="Gsto rozkadu (PDF)")
plt.title("Gsto rozkadu prawdopodobiestwa")
plt.xlabel("Warto")
plt.ylabel("Gsto")
plt.legend()

# Dystrybuanta
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label="Dystrybuanta (CDF)", color="orange")
plt.title("Dystrybuanta rozkadu")
plt.xlabel("Warto")
plt.ylabel("Prawdopodobiestwo")
plt.legend()

plt.tight_layout()
plt.show()

# Obliczanie prawdopodobiestwa, e zmienna losowa osignie warto z zakresu 130 do 145
probability = distribution.cdf(145) - distribution.cdf(130)

# Warto oczekiwana
expected_value = distribution.mean()

# Odchylenie standardowe
standard_deviation = distribution.std()

# Kwantyl rzdu 0.25
quantile_25 = distribution.ppf(0.25)

# Wywietlenie wynikw
results = {
    "Prawdopodobiestwo (130  X  145)": probability,
    "Warto oczekiwana": expected_value,
    "Odchylenie standardowe": standard_deviation,
    "Kwantyl rzdu 0.25": quantile_25,
}

results
```
