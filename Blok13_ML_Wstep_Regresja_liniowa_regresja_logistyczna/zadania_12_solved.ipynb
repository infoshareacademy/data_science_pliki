12.1 Zastosuj one-hot encoding, aby przekształcić poniższy array zawierający informacje na temat preferencji muzycznych na reprezentację binarną.

```python
# Importowanie bibliotek
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# Array z preferencjami muzycznymi
preferencje_muzyczne = np.array(['Rock', 'Pop', 'Jazz', 'Hip-Hop', 'Rock', 'Jazz', 'Pop', 'Hip-Hop']).reshape(-1, 1)

# Tworzenie instancji OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Dopasowanie i przekształcenie danych
encoded_preferences = encoder.fit_transform(preferencje_muzyczne)

# Wyciągnięcie nazw kategorii
categories = encoder.categories_[0]

# Wyświetlenie wyników
encoded_preferences_dict = {
    "Preferencje muzyczne (oryginalne)": preferencje_muzyczne.flatten().tolist(),
    "Kategorie": categories.tolist(),
    "Reprezentacja binarna (one-hot)": encoded_preferences.tolist()
}

# Wyświetlenie słownika z wynikami
encoded_preferences_dict
```

- Napisz funkcjÄ, ktĂłra wyliczaÄ bÄdzie odpowiedĹş funkcji liniowej na podstawie wejĹcia oraz dwĂłch wspĂłĹczynnikĂłw.
- Narysuj wykres funkcji liniowej na podstawie danych syntetycznych.
- StwĂłrz wykres nakĹadajÄcy na siebie kilka funkcji liniowych o rĂłĹźnych wspĂłĹczynnikach slope.
- StwĂłrz wykres nakĹadajÄcy na siebie kilka funkcji liniowych o rĂłĹźnych wspĂłĹczynnikach intercept.
- OdlegĹoĹci miÄdzy dwiema funkcjami liniowymi o rĂłĹźnych wspĂłĹczynnikach slope.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Funkcja obliczajÄca odpowiedĹş funkcji liniowej
def linear_function(x, slope, intercept):
    """
    Oblicza wartoĹÄ funkcji liniowej: y = slope * x + intercept.
    """
    return slope * x + intercept

# Dane syntetyczne dla x
x_values = np.linspace(-10, 10, 100)

# 2. Rysowanie wykresu funkcji liniowej na podstawie danych syntetycznych
y_values = linear_function(x_values, slope=2, intercept=1)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label="y = 2x + 1", color="blue")
plt.title("Wykres funkcji liniowej")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

# 3. Wykres kilku funkcji liniowych o rĂłĹźnych wspĂłĹczynnikach slope
slopes = [0.5, 1, 2, 3]
plt.figure(figsize=(10, 6))

for slope in slopes:
    y_values = linear_function(x_values, slope=slope, intercept=0)
    plt.plot(x_values, y_values, label=f"y = {slope}x")

plt.title("Kilka funkcji liniowych o rĂłĹźnych wspĂłĹczynnikach slope")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

# 4. Wykres kilku funkcji liniowych o rĂłĹźnych wspĂłĹczynnikach intercept
intercepts = [-5, 0, 5, 10]
plt.figure(figsize=(10, 6))

for intercept in intercepts:
    y_values = linear_function(x_values, slope=1, intercept=intercept)
    plt.plot(x_values, y_values, label=f"y = x + {intercept}")

plt.title("Kilka funkcji liniowych o rĂłĹźnych wspĂłĹczynnikach intercept")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

# 5. Obliczanie odlegĹoĹci miÄdzy dwiema funkcjami liniowymi o rĂłĹźnych wspĂłĹczynnikach slope
slope1, slope2 = 1, 2
intercept1, intercept2 = 0, 0

y1 = linear_function(x_values, slope=slope1, intercept=intercept1)
y2 = linear_function(x_values, slope=slope2, intercept=intercept2)

distance = np.abs(y1 - y2)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y1, label="y1 = x", color="blue")
plt.plot(x_values, y2, label="y2 = 2x", color="green")
plt.plot(x_values, distance, label="OdlegĹoĹÄ miÄdzy funkcjami", color="red", linestyle="--")
plt.title("OdlegĹoĹci miÄdzy dwiema funkcjami liniowymi")
plt.xlabel("x")
plt.ylabel("y / OdlegĹoĹÄ")
plt.legend()
plt.grid()
plt.show()
```

12.5
- Narysuj wykres wysokoĹci w czasie.
- Z pierwszego i ostatniego punktu wyznacz rĂłwnanie prostej na podstawie poniĹźszych wzorĂłw.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

df = pd.read_csv("height.csv")

df.head()
df.info()

heightArr = df['Height'].to_numpy()
heightArr

plt.plot(heightArr, linestyle='none', marker='o')

A = (0, heightArr[0])
B = (heightArr.size-1, heightArr[heightArr.size-1])
print(f'A = {A}, B = {B}')

#liczymy wspĂłĹczynniki a i b
a = (B[1]-A[1])/(B[0]-A[0])
b = A[1]-a*A[0]
print("a = {0:.3f}, b = {1}".format(a,b))

def linear_function(a,b,x):
    """
    nasza funkcja liniowa
    """
    return a*x+b

#rysujemy wykres funkcji
fig = plt.figure(figsize=(10,10))
xDraw = np.arange(math.floor(A[0])-1, math.ceil(B[0])+2)
yDraw = linear_function(a,b,xDraw)
plt.plot(xDraw, yDraw)
plt.plot(heightArr, linestyle="none", marker='o')
plt.plot([A[0], B[0]], [A[1], B[1]], linestyle="none", marker='o', markersize=6)

y = heightArr
x = np.arange(0, heightArr.size, 1)
print("x.shape", x.shape)
print("y.shape", y.shape)

x_mean = x.mean()

y_mean = y.mean()

x_sub = x-x_mean

y_sub = y-y_mean

x_sub_times_y_sub = x_sub*y_sub

#x_sub_sq = x_sub**2
x_sub_sq = np.square(x_sub)

xy_sum = x_sub_times_y_sub.sum()

x_sum = x_sub_sq.sum()

a = xy_sum/x_sum

b = y_mean - a*x_mean

#rysujemy wykres funkcji
fig = plt.figure(figsize=(10,10))
xDraw = np.arange(math.floor(A[0])-1, math.ceil(B[0])+2)
yDraw = linear_function(a,b,xDraw)
plt.plot(xDraw, yDraw)
plt.plot(heightArr, linestyle="none", marker='o')

y_ref = y
y_pred = linear_function(a,b,x)

def r_square(y_pred, y_ref, y_mean):
    up = np.sum(np.square(y_ref-y_mean))
    down = np.sum(np.square(y_pred-y_mean))
    return up/down

print("R^2 dla modelu regresji liniowej:\n", r_square(y_pred, y_ref, y_mean))

# sklearn
from sklearn.metrics import r2_score

r2_score(y_ref, y_pred)
```

12.6 Wyznacz model regresji liniowej dla zestawienia "Y house price of unit area" od "X2 house age" w datasecie "estate.csv". Zbadaj poziom korelacji za pomocÄ metryki r2.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.read_csv('estate.csv')

df.head()

df.columns

for k in df.keys():
    tmpNum = df.columns.get_loc(k)
    print(tmpNum,k)

tmpDF = df.sort_values(by=['X2 house age'])

x = np.array(tmpDF.iloc[:,3])
y = np.array(tmpDF.iloc[:,7])
fig = plt.figure(figsize=(10,10))
plt.plot(x, y, linestyle='none', marker='o')
plt.legend()

#przygotowanie danych
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
#model
model = LinearRegression()
#uczenie
model.fit(x,y)

model.score(x, y)

a = model.coef_[0]
b = model.intercept_[0]

def linearFunction(a,b,x):
    return a*x+b

x_reg = x
y_reg = linearFunction(a,b,x_reg)

fig = plt.figure(figsize=(10,10))
plt.plot(x,y,linestyle="none", marker='o', markersize=8, label="real points")
plt.plot(x_reg, y_reg, label="regression model")
plt.legend()

from sklearn.metrics import r2_score
print("R kwadrat: ", r2_score(y, y_reg))
```
